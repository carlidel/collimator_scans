```python
import numpy as np
import matplotlib.pyplot as plt

import fit_functions
import nekhoroshev_tools as nt
from tqdm import tqdm
```

```python
x = np.linspace(1e0, 1e2, 20)
I_1 = 4.0
I_2 = 4.2

I_star = 5.0
exponent = 1 / (2 * 0.33)
c = 1.0
```

```python
for my_c in tqdm(np.linspace(0.1, 10, 5)):
    module = nt.stationary_dist(I_1, I_2, I_star, exponent, my_c) * 2
    d1 = np.asarray(nt.current_generic(
        x,
        lambda p: fit_functions.forward_dist(p, module, I_1, I_2),
        I_2,
        (I_2 / 3) * 2,
        I_star,
        exponent,
        my_c,
    ))


    plt.plot(x, d1, label="d1")
```

```python
for my_I in tqdm(np.linspace(0.1, 10, 5)):
    module = nt.stationary_dist(I_1, I_2, my_I, exponent, c) * 2
    d1 = np.asarray(nt.current_generic(
        x,
        lambda p: fit_functions.forward_dist(p, module, I_1, I_2),
        I_2,
        (I_2 / 3) * 2,
        my_I,
        exponent,
        c,
    ))


    plt.plot(x, d1, label="d1")
```

```python
for k in tqdm(np.linspace(0.05, 0.2, 5)):
    my_exponent = 1 / (2 * k)
    module = nt.stationary_dist(I_1, I_2, I_star*0.01, my_exponent, c) * 2
    d1 = np.asarray(nt.current_generic(
        x,
        lambda p: fit_functions.forward_dist(p, module, I_1, I_2),
        I_2,
        (I_2 / 3) * 2,
        I_star*0.01,
        my_exponent,
        c,
    ))


    plt.plot(x, d1, label="d1")
```
